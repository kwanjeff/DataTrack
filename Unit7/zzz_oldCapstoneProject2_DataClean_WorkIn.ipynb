```python
## Task 1
##import pandas
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
import numpy as np
import os


##import inforamtion for world back of population
#with open(r'D:\springboard\DataScienceGuidedCapstone\Unit7\DataCapstoneProject2\data_worldbank_population.csv') as birthratecsvfile:
LaborRate = pd.read_csv(r'D:\springboard\DataScienceGuidedCapstone\Unit7\DataCapstoneProject2\AllLabour_1960_2020.csv',header=0)
#All Population rate
AllPopulation = pd.read_csv(r'D:\springboard\DataScienceGuidedCapstone\Unit7\DataCapstoneProject2\TotalPopulation.csv',header=0)
#All GDP,  
GDP = pd.read_csv(r'D:\springboard\DataScienceGuidedCapstone\Unit7\DataCapstoneProject2\GDP.csv',header=0)
```

```python
#Task 13
#convert column Japan and United States data to float
LaborRateT.Japan=LaborRateT.Japan.astype('float')
LaborRateT['United States']=LaborRateT['United States'].astype('float')
```

```python
#Task 14
#graph i will be using

rcParams['figure.figsize'] = 30,6
y = LaborRateT['Japan']
y2 = LaborRateT['United States']

x = LaborRateT.index
fig = plt.figure()
ax = plt.subplot(111)
ax.plot(x, y, label='Japan')
ax.plot(x, y2, label='United States')
plt.title('Legend inside')
ax.legend()
```

```python
## Task 1
##import pandas
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
import numpy as np
import os


##import inforamtion for world back of population
#with open(r'D:\springboard\DataScienceGuidedCapstone\Unit7\DataCapstoneProject2\data_worldbank_population.csv') as birthratecsvfile:
LaborRate = pd.read_csv(r'D:\springboard\DataScienceGuidedCapstone\Unit7\DataCapstoneProject2\AllLabour_1960_2020.csv',header=0)
```

```python
# Task 5
#create new df to move data for japan and United states into
LaborRate2 = LaborRate.loc[LaborRate['CountryName'].isin(['Japan','United States'])]
LaborRate2
```

```python
# Task 6
LaborRate2.columns
```

```python
# Task 7
#drop CountryCode column as it doesnt provide useful data or needed in mapping
LaborRate2= LaborRate2.drop(['CountryCode'],axis = 1)
LaborRate2
```

```python
# Task 8
#Rename Column name of CountryName to Year
LaborRate2.rename(columns={'CountryName':'Year'},inplace = True)
```

```python
# Task 9
#print out data
LaborRate2
```

```python
# Task 9
#Transpose axis for LaborRate2
LaborRateT = LaborRate2.T
LaborRateT
```

```python
#Task 10
#use to change col to new name
LaborRateT.columns = LaborRateT.iloc[0]
LaborRateT = LaborRateT.drop(LaborRateT.index[0])
```

```python
#Task 11
#use to change col to new name
LaborRateT
```

```python
#Task 12
LaborRateT.dtypes
```

```python
#Task 13
#convert column Japan and United States data to float
LaborRateT.Japan=LaborRateT.Japan.astype('float')
LaborRateT['United States']=LaborRateT['United States'].astype('float')
```

```python
#Task 14
#graph i will be using

rcParams['figure.figsize'] = 30,6
y = LaborRateT['Japan']
y2 = LaborRateT['United States']

x = LaborRateT.index
fig = plt.figure()
ax = plt.subplot(111)
ax.plot(x, y, label='Japan')
ax.plot(x, y2, label='United States')
plt.title('Legend inside')
ax.legend()
```
